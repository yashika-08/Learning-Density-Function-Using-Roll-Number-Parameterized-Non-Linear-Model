```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    r"C:\Users\Yashika\Desktop\DS Assignments\Learning Probability Density Function\data.csv",
    encoding="latin1"
)
# Display the first few rows of the DataFrame
print(df.head())
```

```python
x = df['no2'].dropna().values
```

```python
r = 102303439
a_r = 0.05 * (r % 7)
b_r = 0.3 * ((r % 5) + 1)

print("a_r =", a_r)
print("b_r =", b_r)
```

```python
z = x + a_r * np.sin(b_r * x)
```

```python
mu = np.mean(z)
sigma2 = np.var(z)

lam = 1 / (2 * sigma2)
c = 1 / np.sqrt(2 * np.pi * sigma2)

print("mu =", mu)
print("lambda =", lam)
print("c =", c)
```

```python
z_range = np.linspace(min(z), max(z), 1000)
pdf = c * np.exp(-lam * (z_range - mu)**2)

plt.hist(z, bins=50, density=True, alpha=0.5, label="Histogram of z")
plt.plot(z_range, pdf, 'r', label="Learned PDF")
plt.legend()
plt.title("PDF Learning of Transformed Variable")
plt.show()
```
